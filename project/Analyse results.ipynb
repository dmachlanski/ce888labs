```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels = ['100', '200', '400', '800']
data_evse = []
data_uctvsdt = {'100': [], '200': [], '400': [], '800': []}

for l in labels:
    df = pd.read_csv(f'final_results/results{l}/results_1.csv')
    df['Score'] = df['Wins'] + (df['Draws'] * 0.5)
    data_evse.append(df)
    for l2 in labels:
        df2 = pd.read_csv(f'final_results/results{l}/results_2_{l2}.csv')
        df2['Score'] = df2['Wins'] + (df2['Draws'] * 0.5)
        data_uctvsdt[l].append(df2)
```

```python
data_uctvsdt['100'][3]
```

```python
iters = np.arange(10, (len(data_evse[0])+1)*10, 10)
```

```python
styles = ['bs--', 'go-', 'r^-.', 'k+:']

for d, s, l in zip(data_evse, styles, labels):
    plt.plot(iters, d['Score'], s, label=l)

plt.axis([10, 200, 40, 70])
plt.xlabel('Iterations')
plt.ylabel('Score')
plt.legend(loc='upper left', title='Itermax:')
plt.savefig('images/e_vs_e.png')
plt.show()
```

```python
# DT against UCT(itermax=100)
for k, s in zip(data_uctvsdt, styles):
    plt.plot(iters, data_uctvsdt[k][0]['Score'], s, label=k)

plt.axis([10, 200, 40, 100])
plt.xlabel('Iterations')
plt.ylabel('Score')
plt.legend(loc='upper left', title='Itermax:')
plt.savefig('images/uct_vs_dt_100.png')
plt.show()
```
